```python
G=nx.Graph()
G.add_edge(1,2,w=0.1)
G.add_edge(2,3,w=0.2)
G.add_edge(3,4,w=0.3)
G.add_edge(4,5,w=0.8)
G.add_edge(5,6,w=0.7)
G.add_edge(1,6,w=0.1)
G.add_edge(1,4,w=0.4)
G.add_edge(2,5,w=0.6)
G.add_edge(3,6,w=0.5)
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,node_color=range(6))
nx.draw_networkx_labels(G,pos)
X=[(m,n) for (m,n,r) in G.edges(data=True) if  r['w']<0.6]
Y=[(m,n) for (m,n,r) in G.edges(data=True) if  r['w']>=0.6]
nx.draw_networkx_edges(G,pos,edgelist=X, edge_color='r',style='dashed')
nx.draw_networkx_edges(G,pos,edgelist=Y ,edge_color='g',width=6)
```

```python
G=nx.Graph()
e=[(1,2,1), (2,3,2),(3,4,3),(4,5,8),(5,6,7),(1,6,1),(1,4,4),(2,5,6),(3,6,5)]
G.add_weighted_edges_from(e)

pos=nx.spring_layout(G)
nx.draw(G,pos,labels={n: n for n in G.nodes()})

l={(1,2):1, (2,3):2,(3,4):3,(4,5):8,(5,6):7,(1,6):1,(1,4):4,(2,5):6,(3,6):5}
nx.draw_networkx_edge_labels(G,pos,edge_labels=l,font_color='b')
```
