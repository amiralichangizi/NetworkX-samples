```python
WG=nx.Graph()
e=[(1,2,5),(2,3,10),(3,1,15),(3,4,20)]
WG.add_weighted_edges_from(e)
nx.draw_networkx(WG)
```

```python
WG=nx.Graph()
e=[(1,2,5),(2,3,10),(3,1,15),(3,4,20),(2,4,16)]
WG.add_weighted_edges_from(e)

pos=nx.spring_layout(WG)
nx.draw(WG,pos,labels={n:n for n in WG.nodes()})

l={(1,2):5, (2,3):10, (3,1):15, (3,4):20,(2,4):16}
nx.draw_networkx_edge_labels(WG,pos,edge_labels=l);
```
