```python
n=8
G=nx.star_graph(n)
pos=nx.spring_layout(G)
nx.draw(G,node_color=range(n+1),width=5, edge_color=range(n))
```

```python
G=nx.Graph()
G.add_edge(1,2,w=0.1)
G.add_edge(2,3,w=0.2)
G.add_edge(3,4,w=0.3)
G.add_edge(4,5,w=0.8)
G.add_edge(5,6,w=0.7)
G.add_edge(1,6,w=0.1)
G.add_edge(1,4,w=0.4)
G.add_edge(2,5,w=0.6)
G.add_edge(3,6,w=0.5)
pos=nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos,node_color=range(6))
nx.draw_networkx_labels(G,pos)
X=[(m,n) for (m,n,r) in G.edges(data=True) if  r['w']<0.3]
Y=[(m,n) for (m,n,r) in G.edges(data=True) if  r['w']>=0.3]
nx.draw_networkx_edges(G,pos,edgelist=X, edge_color='r',style='dashed')
nx.draw_networkx_edges(G,pos,edgelist=Y ,edge_color='g',width=6)

```
